## Null Value Identification

```python

def check_nulls(df):
    df.select([
        count(when(col(c).isNull(), c)).alias(c + "_nulls")
        for c in df.columns
    ]).show()

check_nulls(customers_df)
check_nulls(products_df)
check_nulls(categories_df)
check_nulls(subcategories_df)
check_nulls(sales_2015_df)
check_nulls(sales_2016_df)
check_nulls(sales_2017_df)
check_nulls(returns_df)
check_nulls(calendar_df)
check_nulls(territories_df)
```

## Negative Value Detection in Numeric Columns

```python
def check_negatives(df):
    numeric_cols = [c for c, t in df.dtypes if t in ['int', 'double', 'float', 'long']]
    
    for c in numeric_cols:
        print(f"Negative values in {c}")
        df.filter(col(c) < 0).show()

check_negatives(products_df)
check_negatives(sales_2015_df)
check_negatives(sales_2016_df)
check_negatives(sales_2017_df)
check_negatives(returns_df)
```

## Negative Value Count per Column

```python
def count_negatives(df):
    numeric_cols = [c for c, t in df.dtypes if t in ['int', 'double', 'float', 'long']]
    
    df.select([
        count(when(col(c) < 0, c)).alias(c + "_negatives")
        for c in numeric_cols
    ]).show()

count_negatives(products_df)
count_negatives(sales_2015_df)
count_negatives(sales_2016_df)
count_negatives(sales_2017_df)
count_negatives(returns_df)
```

## Duplicate Record Identification

```python
customers_df.groupBy("CustomerKey").count().filter(col("count") > 1).show()

products_df.groupBy("ProductKey").count().filter(col("count") > 1).show()

categories_df.groupBy("ProductCategoryKey").count().filter(col("count") > 1).show()

subcategories_df.groupBy("ProductSubcategoryKey").count().filter(col("count") > 1).show()

sales_2015_df.groupBy("OrderNumber").count().filter(col("count") > 1).show()
sales_2016_df.groupBy("OrderNumber").count().filter(col("count") > 1).show()
sales_2017_df.groupBy("OrderNumber").count().filter(col("count") > 1).show()

returns_df.groupBy("ReturnDate", "TerritoryKey", "ProductKey").count().filter(col("count") > 1).show()

calendar_df.groupBy("Date").count().filter(col("count") > 1).show()

territories_df.groupBy("SalesTerritoryKey").count().filter(col("count") > 1).show()
```

## Invalid String Value & Empty String Detection

```python
def check_invalid_strings(df):
    string_cols = [c for c, t in df.dtypes if t == 'string']
    
    for c in string_cols:
        print(f"Invalid values in {c}")
        df.filter(
            (trim(col(c)) == "") |
            (lower(col(c)).isin("na", "null"))
        ).show()

check_invalid_strings(customers_df)
check_invalid_strings(products_df)
check_invalid_strings(categories_df)
check_invalid_strings(subcategories_df)
check_invalid_strings(territories_df)
```
